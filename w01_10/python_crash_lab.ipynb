# JSON and API Basic

```python
import json
# JSON-like structure
agent_data = {
    "agent": "Alpha",
    "role": "assistant",
    "goals": ["plan", "analyze", "act"]
}

#Convert JSON to string
json_str = json.dumps(agent_data)
print ("JSON string:", json_str)

#Back to Python Dictionary 
parsed= json.loads(json_str)
print("Agent Goals:", parsed["goals"])

print("Agnent role:", parsed["role"])
```
